```python
from pylab import * 
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
import pandas as pd
```

```python
df = pd.read_excel('kraftpris.xlsx')
df["Tid"]
df["Tid"] = df["Tid"].replace(-9999, np.nan).dropna().astype(int)
df["Kraftpris"].mean()
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(16,8))

ax.plot(df["Tid"], df["Kraftpris"], color="blue", lw=1, ls='-', marker='s', markersize=6, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");

#ax.axvspan(0, 17, ymin=0.5, ymax=1, alpha=0.1, color='blue')

ax.plot([2000,2017],[33.5,33.5],color="red", lw=2, ls='--')

NVEkraftpris = [27.5,43,39,32,36,37,39]
NVEar = [2017,2018,2019,2020,2023,2025,2030]

ax.plot(NVEar,NVEkraftpris,color="black", lw=2, ls='--')

xs = [2000,2030]
ys = [33,33]
ym = [45,45]
ax.set_ylim(0,70)
ax.set_xlim(1999,2030)
ax.fill_between(xs, ys,ym, interpolate=True,alpha =0.2, color='blue')

xv = [2000,2005,2010,2015,2020,2025,2030]
yvs = [60,53,47,40,35,29,27]
yvm = [75,67,58,52,48,42,41]
ax.fill_between(xv, yvs,yvm, interpolate=True,alpha =0.2, color='green')

ax.legend(['Kraftpris 2000-2017','Gjennomsnittlig kraftpris 2000-2017','NVE-prognose 2018-2030','LCOE vannkraft','LCOE vindkraft'])
fig.suptitle("Oversikt over utvikling av krafrpriser, prognoser og investeringskostnader", fontsize=16)
ax.set_ylabel('øre/kWh',fontsize = 16)
plt.grid(True)
```
